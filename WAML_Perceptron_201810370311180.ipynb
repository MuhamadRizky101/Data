```python
class Perceptron(object):
    
    def __init__(self, num_of_inputs, epochs=10, learning_rate=1):
        self.epochs = epochs
        self.learning_rate = learning_rate
        self.weights = np.zeros(num_of_inputs + 1)
    
    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        if summation >= 0:
            activation = 1
        else:
            activation = 0
        print('output', activation)
        return activation
    
    def train(self, training_inputs, labels):
        i = 0
        for _ in range(self.epochs):
            i += 1
            print('###EPOCHS###', i)
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                self.weights[1:] = self.weights[1:] + self.learning_rate * (label - prediction) * inputs
                self.weights[0] = self.weights[0] + self.learning_rate * (label - prediction)
                print('label', label)
                print('weight', self.weights[1:])
                print('bias', self.weights[0])
                print(' ') 
```

```python
# Data
training_inputs = []
for dt_X_train in np.array(x_train):
  training_inputs.append (np.array(dt_X_train))

labels = np.array(y_train)

# Training
perceptron = Perceptron(2)
perceptron.train(training_inputs, labels)
```

```python
# Prediksi menggunakan data train

data_compare_train = []
for data in np.array(x_train):
  data_compare_train.append(perceptron.predict(data))

n = 0
data_labels = list (y_train)
for x in range (len(data_compare_train)):
  if data_compare_train[x] == data_labels[x] :
    n+=1

accuracy = (n/len(data_compare_train))*100
print ('\nAccuracy :', accuracy)
```
